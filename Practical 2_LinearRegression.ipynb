Removing Outliers from price

```python
meanA = data['price'].mean()
Q1 = np.percentile(data['price'],25)
Q2 = np.median(data['price'])
Q3 = np.percentile(data['price'],75)
IQR = Q3 - Q1
lower_fence = Q1 - (1.5*IQR)
upper_fence = Q3 + (1.5*IQR)

print("lower fence is " +str(lower_fence))
print("upper fence is " +str(upper_fence))
```

```python
data = data.loc[(data['price']<= upper_fence) & (data['price']>=lower_fence)]
data
```

```python
sns.boxplot(data.price)
```

```python
x = data.drop(['stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea',
               'furnishingstatus'], axis = 1)
```

```python
x
```

```python
x = x.drop('price',axis = 1)
```

```python
x
```

```python
y = data.price
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, train_size = 0.7, random_state = 100)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test,y_pred)
r_squared = r2_score(y_test,y_pred)
r_squared
```

```python
from math import sqrt
rmse = sqrt(mse)
print('Mean_Squared_Error :' ,mse)
print('Root_Mean_Squared_Error :' ,rmse)
print('r_square_value :',r_squared)
```
